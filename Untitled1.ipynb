```python
import matplotlib.pyplot as plt

# Configuramos una figura con 2 subplots (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# --- GRÁFICO 1: COMPARATIVA FACTURACIÓN ($) ---
# Eje principal (Azul) - Empresa Total
ax1.plot(resumen_mensual.index, resumen_mensual['FACTURACION  TOTAL'], color='tab:blue', marker='o', linewidth=2, label='Total Empresa')
ax1.set_ylabel('Facturación Empresa ($)', color='tab:blue', fontsize=11)
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Eje secundario (Rojo) - PC Gamer (usamos twinx porque las escalas son muy distintas)
ax1_twin = ax1.twinx()
ax1_twin.plot(resumen_gamer.index, resumen_gamer['FACTURACION  TOTAL'], color='tab:red', marker='s', linestyle='--', label='PC Gamer (5527)')
ax1_twin.set_ylabel('Facturación PC Gamer ($)', color='tab:red', fontsize=11)
ax1_twin.tick_params(axis='y', labelcolor='tab:red')

ax1.set_title('Tendencia de Facturación: Total vs SKU 5527', fontsize=13, fontweight='bold')
ax1.set_xticks(resumen_mensual.index)
ax1.set_xticklabels(['Dic 2025', 'Ene 2026', 'Feb 2026'])
ax1.grid(True, alpha=0.3)

# --- GRÁFICO 2: COMPARATIVA UNIDADES VENDIDAS ---
# Eje principal (Verde) - Empresa Total
ax2.plot(resumen_mensual.index, resumen_mensual['VENTAS'], color='tab:green', marker='o', linewidth=2, label='Total Empresa')
ax2.set_ylabel('Unidades Empresa', color='tab:green', fontsize=11)
ax2.tick_params(axis='y', labelcolor='tab:green')

# Eje secundario (Naranja) - PC Gamer
ax2_twin = ax2.twinx()
ax2_twin.plot(resumen_gamer.index, resumen_gamer['VENTAS'], color='tab:orange', marker='s', linestyle='--', label='PC Gamer (5527)')
ax2_twin.set_ylabel('Unidades PC Gamer', color='tab:orange', fontsize=11)
ax2_twin.tick_params(axis='y', labelcolor='tab:orange')

ax2.set_title('Tendencia de Unidades: Total vs SKU 5527', fontsize=13, fontweight='bold')
ax2.set_xticks(resumen_mensual.index)
ax2.set_xticklabels(['Dic 2025', 'Ene 2026', 'Feb 2026'])
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
